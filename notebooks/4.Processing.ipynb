## Limpeza do dataset

```python
# Outlier: casas muito grandes distorcem a regressão
# Aplicado APENAS no treino
X_train = X_train[X_train["Gr Liv Area"] < 4000].copy()
y_train = y_train[X_train.index]  # manter índices sincronizados

# Erro tipográfico conhecido — corrigir em ambos
# (não é uma decisão estatística, é correção de dado)
for df_ in [X_train, X_test]:
    df_["Garage Yr Blt"] = df_["Garage Yr Blt"].replace(2207, 2007)

print(f"Treino após remoção de outliers: {X_train.shape[0]} registros")
print(f"Teste intocado                 : {X_test.shape[0]} registros")
```

### Feature engeneering de para variáveis que serão excluídas

```python
# Criar features compostas ANTES de dropar as colunas originais
for df_ in [X_train, X_test]:
    df_["total_bath"] = (
        df_["Bsmt Full Bath"].fillna(0) * 1.0 +
        df_["Bsmt Half Bath"].fillna(0) * 0.5 +
        df_["Full Bath"].fillna(0)      * 1.0 +
        df_["Half Bath"].fillna(0)      * 0.5
    )
    df_["total_outdoor_sf"] = (
        df_["Wood Deck SF"].fillna(0) +
        df_["Open Porch SF"].fillna(0) +
        df_["Enclosed Porch"].fillna(0)
    )

cols_to_drop = [
    # Baixa variância
    "Street", "Land Contour", "Utilities", "Lot Config",
    "Land Slope", "Condition 2", "Roof Matl", "Heating",
    "Electrical", "Kitchen AbvGr", "Functional",
    "3Ssn Porch", "Screen Porch",

    # Excesso de nulos / raridade
    "Alley", "Pool Area", "Pool QC", "Fence",
    "Misc Feature", "Misc Val",

    # Semântica / redundância de design
    "Lot Shape", "Condition 1", "Bldg Type",
    "Roof Style", "Fireplace Qu", "Paved Drive",
    "Mo Sold", "Yr Sold",

    # VIF alto / redundância numérica
    "Exterior 1st", "Exterior 2nd",
    "BsmtFin Type 1", "BsmtFin SF 1",
    "BsmtFin Type 2", "BsmtFin SF 2",
    "Bsmt Unf SF",
    "1st Flr SF", "2nd Flr SF", "Low Qual Fin SF",
    "TotRms AbvGrd", "Garage Area",

    # Já agregados em total_bath
    "Bsmt Full Bath", "Bsmt Half Bath",
    "Full Bath", "Half Bath",

    # Já agregados em total_outdoor_sf
    "Wood Deck SF", "Open Porch SF", "Enclosed Porch",

    "Sale Type", "Sale Condition"
]

cols_to_drop = [c for c in cols_to_drop if c in X_train.columns]

for df_ in [X_train, X_test]:
    df_.drop(columns=cols_to_drop, inplace=True)

print(f"Colunas restantes: {X_train.shape[1]}")
```

### Imputs

```python
# Lot Frontage: mediana por bairro
lot_medians = X_train.groupby("Neighborhood")["Lot Frontage"].median()
lot_global  = X_train["Lot Frontage"].median()  # fallback

for df_ in [X_train, X_test]:
    df_["Lot Frontage"] = df_["Lot Frontage"].fillna(
        df_["Neighborhood"].map(lot_medians)
    ).fillna(lot_global)  # bairros não vistos no treino

# Mas Vnr Area: zero se ausente (sem fit)
for df_ in [X_train, X_test]:
    df_["Mas Vnr Area"] = df_["Mas Vnr Area"].fillna(0)

# Ordinais de porão e garagem: "None" = ausente
cat_fill_none = [
    "Bsmt Qual", "Bsmt Cond", "Bsmt Exposure",
    "Garage Qual", "Garage Cond", "Garage Finish",
    "Mas Vnr Type",
]
for df_ in [X_train, X_test]:
    for col in cat_fill_none:
        if col in df_.columns:
            df_[col] = df_[col].fillna("None")

# Garage Yr Blt: fallback para Year Built (sem fit)
for df_ in [X_train, X_test]:
    mask = df_["Garage Yr Blt"].isna()
    df_.loc[mask, "Garage Yr Blt"] = df_.loc[mask, "Year Built"]

for df_ in [X_train, X_test]:
    # Imóveis sem porão
    df_["Total Bsmt SF"] = df_["Total Bsmt SF"].fillna(0)
    
    # Imóveis sem garagem
    df_["Garage Cars"] = df_["Garage Cars"].fillna(0)
    df_["Garage Type"] = df_["Garage Type"].fillna("None")

print("Nulos treino:", X_train.isnull().sum().sum())
print("Nulos teste :", X_test.isnull().sum().sum())
```

### Feature engeneering

```python
REF_YEAR = 2010

# Idades (determinístico — sem fit)
for df_ in [X_train, X_test]:
    df_["house_age"]  = REF_YEAR - df_["Year Built"]
    df_["remod_age"]  = REF_YEAR - df_["Year Remod/Add"]
    df_["garage_age"] = REF_YEAR - df_["Garage Yr Blt"]
    df_.drop(columns=["Year Built", "Year Remod/Add", "Garage Yr Blt"],
             inplace=True)

# price_tier: quartis calculados APENAS no treino
nbhd_medians_train = y_train.groupby(X_train["Neighborhood"]).median()
q33 = nbhd_medians_train.quantile(0.33)
q66 = nbhd_medians_train.quantile(0.66)

def assign_tier(nbhd):
    val = nbhd_medians_train.get(nbhd, None)
    if val is None:   return "mid"   # bairro não visto no treino
    if val <= q33:    return "low"
    elif val <= q66:  return "mid"
    else:             return "high"

X_train["price_tier"] = X_train["Neighborhood"].map(assign_tier)
X_test["price_tier"]  = X_test["Neighborhood"].map(assign_tier)

for df_ in [X_train, X_test]:
    df_.drop(columns=["Neighborhood"], inplace=True)

# Interação qualidade × condição (determinístico — sem fit)
for df_ in [X_train, X_test]:
    df_["qual_cond_interaction"] = df_["Overall Qual"] * df_["Overall Cond"]

# Central Air → binário (determinístico — sem fit)
for df_ in [X_train, X_test]:
    df_["Central Air"] = (df_["Central Air"] == "Y").astype(int)
```

### Encoding das variaveis categoricas ordinais

```python
QUALITY_MAP       = {"Ex": 5, "Gd": 4, "TA": 3, "Fa": 2, "Po": 1, "None": 0}
BSMT_EXPOSURE_MAP = {"Gd": 4, "Av": 3, "Mn": 2, "No": 1, "None": 0}
GARAGE_FINISH_MAP = {"Fin": 3, "RFn": 2, "Unf": 1, "None": 0}

ordinal_mappings = {
    "Exter Qual"    : QUALITY_MAP,
    "Exter Cond"    : QUALITY_MAP,
    "Bsmt Qual"     : QUALITY_MAP,
    "Bsmt Cond"     : QUALITY_MAP,
    "Bsmt Exposure" : BSMT_EXPOSURE_MAP,
    "Heating QC"    : QUALITY_MAP,
    "Kitchen Qual"  : QUALITY_MAP,
    "Garage Qual"   : QUALITY_MAP,
    "Garage Cond"   : QUALITY_MAP,
    "Garage Finish" : GARAGE_FINISH_MAP,
}

for df_ in [X_train, X_test]:
    for col, mapping in ordinal_mappings.items():
        if col in df_.columns:
            df_[col] = df_[col].map(mapping).fillna(0).astype(int)
```

### One-Hot-Encoding

```python
RARE_ZONES = ["RH", "RM", "C (all)"]
COMMON_VNR = ["BrkFace", "Stone", "None"]

for df_ in [X_train, X_test]:
    df_["MS Zoning"] = df_["MS Zoning"].apply(
        lambda z: "Other" if z in RARE_ZONES else z
    )
    df_["Mas Vnr Type"] = df_["Mas Vnr Type"].apply(
        lambda v: v if v in COMMON_VNR else "Other"
    )

one_hot_cols = [
    "MS SubClass", "MS Zoning", "House Style",
    "Mas Vnr Type", "Foundation", "Garage Type", "price_tier",
]

# Fit apenas no treino
X_train = pd.get_dummies(X_train, columns=one_hot_cols,
                         drop_first= True, dtype=int)

# Apply no teste e alinhar colunas com o treino
X_test = pd.get_dummies(X_test, columns=one_hot_cols,
                        drop_first=True, dtype=int)

X_test = X_test.reindex(columns=X_train.columns, fill_value=0)

assert X_train.shape[1] == X_test.shape[1], "Colunas divergem!"
print(f"Features finais: {X_train.shape[1]}")
```

### Validação

```python
def reverse_one_hot(df):
    # Identifica qual coluna price_tier_... tem o valor 1 em cada linha
    # Nota: Se houver uma classe 'high' que foi dropada no get_dummies, 
    # ela será identificada onde todas as outras são 0.
    
    conditions = [
        (df['price_tier_low'] == 1),
        (df['price_tier_mid'] == 1)
    ]
    choices = ['Baixo', 'Médio']
    
    # Criamos a coluna preenchendo com 'Alto' onde low e mid são 0
    df['price_tier'] = np.select(conditions, choices, default='Alto')
    return df

# Aplicando aos seus datasets
treino_final = reverse_one_hot(X_train)
teste = reverse_one_hot(X_test)
```
