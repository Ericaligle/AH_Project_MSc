```python
summary_table = pd.DataFrame({
    'Num of Values': df.count(),
    '% of Missing Values': df.isnull().mean() * 100,
    'Unique Values (Categorical)': df.select_dtypes(include=['object', 'string']).nunique(),
    'Mean (Numerical)': df.select_dtypes(include='number').mean(),
    'Python Type': df.dtypes,
})

pd.set_option('display.max_rows', None)
summary_table
```

## Missing Values

```python
varibles_with_missing_values = summary_table[summary_table['% of Missing Values'] > 0].T.columns
varibles_with_missing_values
```

```python
for col in varibles_with_missing_values:
    print(f"{col}:\n {df[col].isnull().sum()} missing values")
    print(f"Percentage: {df[col].isnull().mean() * 100:.2f}%")
    print(f'{df[col].value_counts(dropna=False)}\n')
    print("-" * 40)
```

```python
# Verificando valores faltantes da coluna'pool qc' em relação a 'pool area'

df[df['Pool QC'].isnull()][['Pool Area', 'Pool QC']].value_counts(dropna=False)
```

```python
''' Verificando valores faltantes da coluna 'Mas Vnr Type' em relação a 'Mas Vnr Area'
 Observando que existem casos onde 'Mas Vnr Area' é maior ou igual a 0, mas 'Mas Vnr Type'
 está faltando, o que pode indicar uma inconsistência nos dados.'''

df.loc[(df['Mas Vnr Type'].isnull()) & (df['Mas Vnr Area'] >= 0), ['Mas Vnr Area', 'Mas Vnr Type']].value_counts(dropna=False)
```

```python
''' Verificando valores faltantes da coluna 'Misc Feature' em relação a 'Misc Val'
 Observando que existem casos onde 'Misc Val' é igual a 0, mas 'Misc Feature' está null,
 Pode indicar uma inconsistência nos dados.'''

df[df['Misc Feature'].isnull()][['Misc Feature', 'Misc Val']].value_counts(dropna=False)
```

```python
''' Verificando valores faltantes da coluna 'Fireplace Qu' em relação a 'Fireplaces'
Não há retorno de valores onde a quantidade de lareiras seja diferente de 0, o que indica que
Fireplace Qu está dentro do esperado (não existe lareira na propriedade)
'''

df[(df['Fireplace Qu'].isnull()) & (df['Fireplaces'] != 0)]
```

```python
''' Verificando valores faltantes da coluna 'Lot Frontage' em relação a 'Lot Shape'
 Observando a possibilidade de que os valores faltantes ou negativos em 'Lot Frontage'
possam estar relacionados a diferentes formas de lote, o que pode indicar uma inconsistência nos dados.
de 480 casos onde 'Lot Frontage' é null ou menor ou igual a 0, 358 estão relacionados a lotes com formato irregular (Lot Shape = IR1, IR2, IR3)
'''

df.loc[(df['Lot Frontage'].isnull()) | (df['Lot Frontage'] <= 0), ['Lot Frontage', 'Lot Shape']].groupby('Lot Shape').size().sort_values(ascending=False)
```

```python
''' Verificando por que os valores faltantes da coluna 'Garage Type' não estão de acordo com 'Garage Yr Blt'
'''

df.loc[(df['Garage Yr Blt'].isnull()), ['Garage Type', 'Garage Yr Blt', ]].value_counts(dropna=False)
```

```python
''' No índice 1356, 'Garage Type' é 'Detchd', possui quantidade de carros suportados e área de garagem.
Provavelmente essa garagem existe, mas houve falha na coleta das outras informações, como o ano de construção,
acababento, qualidade e condição

Já no índice 2236, 'Garage Type' é 'Detchd', mas não possui quantidade de carros suportados e área de garagem,
o que pode indicar que a garagem não existe ou que houve erro no preenchimento da informação.'''

df.loc[(df['Garage Type'] == 'Detchd') & (df['Garage Yr Blt'].isnull())]
```
