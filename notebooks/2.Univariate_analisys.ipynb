## Análise univariada

### Variáveis numéricas

```python
baixa_variabilidade = []
alta_variabilidade = []
moderada_variabilidade = []
skewness_high = []
skewness_moderate = []
skewness_low = []
kurtosis_high = []
kurtosis_moderate = [] 
kurtosis_low = []


for col in numerical_cols:
    print(f"===== Coluna {col} =====\n")

    print(f"{df[col].describe()}\n")

    cv = df[col].std() / df[col].mean()

    if cv < 0.3:
        print(f"Coeficiente de Variação: CV = {cv:.2f} » Baixa variabilidade - valores concentrados em torno da média")
        baixa_variabilidade.append(col)
    if cv >= 0.7:
        print(f"Coeficiente de Variação: CV = {cv:.2f} » Alta variabilidade - valores mais dispersos em relação à média")
        alta_variabilidade.append(col)
    if 0.3 <= cv < 0.7:
        print(f"Coeficiente de Variação: CV = {cv:.2f} » Variabilidade moderada - pode indicar assimetria dos dados e/ou presença de outliers")
        moderada_variabilidade.append(col)

    iqr = df[col].quantile(0.75) - df[col].quantile(0.25)
    print(f"Range: {df[col].max() - df[col].min()}")

    print(f"IQR: {iqr}")
    print(f"Limite inferior: {df[col].quantile(0.25) - 1.5 * iqr}")
    print(f"Limite superior: {df[col].quantile(0.75) + 1.5 * iqr}")
    print(f"Quantidade de outliers: {((df[col] < (df[col].quantile(0.25) - 1.5 * iqr)) | (df[col] > (df[col].quantile(0.75) + 1.5 * iqr))).sum()}")

    skewness = df[col].skew()
    kurtosis = df[col].kurtosis()

    if skewness < 0.5:
        print(f"Assimetria: {skewness:.2f} » Approximately symmetric")
        skewness_low.append(col)
    elif 0.5 <= skewness < 1:
        print(f"Assimetria: {skewness:.2f} » Moderately skewed")
        skewness_moderate.append(col)
    else:
        print(f"Assimetria: {skewness:.2f} » Highly skewed")
        skewness_high.append(col)

    if kurtosis == 0:
        print(f"Curtose: {kurtosis:.2f} » Similar to normal distribution")
        kurtosis_low.append(col)
    elif 0 < kurtosis < 3:
        print(f"Curtose: {kurtosis:.2f} » Moderately heavy tails")
        kurtosis_moderate.append(col)
    else:
        print(f"Curtose: {kurtosis:.2f} » Heavy tails and more extreme values")
        kurtosis_high.append(col)

    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(6, 3))
    ax1.boxplot(df[col].dropna())
    ax1.set_title(f"Boxplot da coluna {col}")
    ax2.hist(df[col], bins=30, edgecolor='black', )
    ax2.set_title(f"Histograma da coluna {col}")
    fig.suptitle(f"Análise da coluna {col}")
    plt.tight_layout()
    plt.show()

    print("=" * 80)
```

```python
# criar uma função que verifique se cada coluna está presente em cada uma das listas de variabilidade,
# assimetria e curtose, para facilitar a análise posterior
for column in numerical_cols:
    print(f"\n===== Analisando a coluna: {column} ====")
    
    if column in baixa_variabilidade:
        print("Variabilidade: Baixa")
    elif column in moderada_variabilidade:
        print("Variabilidade: Moderada")
    elif column in alta_variabilidade:
        print("Variabilidade: Alta")
    else:
        print("Variabilidade: Não classificada")

    if column in skewness_low:
        print("Assimetria: Baixa (Approximately symmetric)")
    elif column in skewness_moderate:
        print("Assimetria: Moderada (Moderately skewed)")
    elif column in skewness_high:
        print("Assimetria: Alta (Highly skewed)")
    else:
        print("Assimetria: Não classificada")

    if column in kurtosis_low:
        print("Curtose: Baixa (Similar to normal distribution)")
    elif column in kurtosis_moderate:
        print("Curtose: Moderada (Moderately heavy tails)")
    elif column in kurtosis_high:
        print("Curtose: Alta (Heavy tails and more extreme values)")
    else:
        print("Curtose: Não classificada")
```

### Variáveis Categoricas

```python
cols_baixa_variabilidade = []


for col in categorical_cols:
    valores_unicos = df[col].nunique()
    mais_frequente = df[col].mode().iloc[0]
    #frequencia_mais_frequente = df[col].value_counts().iloc[0]
    porcentagem_mais_frequente = df[col].value_counts(normalize=True, dropna=False).iloc[0] * 100
    #menos_frequente = df[col].value_counts().index[-1]
    #frequencia_menos_frequente = df[col].value_counts().iloc[-1]
    porcentagem_menos_frequente = df[col].value_counts(normalize=True, dropna=False).iloc[-1] * 100
    categorias_raras = []
    
    for cat in df[col]:
        try:
            if df[col].value_counts(normalize=True, dropna=False)[cat] < 0.05:
                categorias_raras.append(cat)
        except KeyError:
            pass

    print(f"\n===== {col} =====\n")
    #print(f"Valores únicos: {valores_unicos}")
    #print(f"Valor mais frequente: {mais_frequente}")
    #print(f"Frequência do valor mais frequente: {frequencia_mais_frequente}")
    #print(f"Porcentagem do valor mais frequente: {porcentagem_mais_frequente:.2f}%")
    print(f"Contagem de cada categoria:\n{df[col].value_counts(dropna=False)}\n")
    #print(f"Valor menos frequente: {menos_frequente}")
    #print(f"Frequência do valor menos frequente: {frequencia_menos_frequente}")
    #print(f"Porcentagem do valor menos frequente: {porcentagem_menos_frequente:.2f}%")
    print(f"Categorias raras (menos de 5% dos dados): {set(categorias_raras)}")

    if porcentagem_mais_frequente > 90:
        #print(f"\n===== {col} =====\n")
        cols_baixa_variabilidade.append(col)
        print(f"Coluna pode ter baixo poder discriminatório (a categoria -{mais_frequente}- representa {porcentagem_mais_frequente:.2f}% dos dados)")

    if valores_unicos >= 10 and valores_unicos < 30:
        print(f" tem cardinalidade moderada ({valores_unicos} valores únicos)")
    if valores_unicos >= 30 and valores_unicos < 50:
        print(f" tem cardinalidade alta ({valores_unicos} valores únicos)")
    if valores_unicos >= 50:
        print(f"tem cardinalidade muito alta (Não é um bom preditor para modelos lineares)")
    if valores_unicos < 10:
        print(f"tem cardinalidade baixa ({valores_unicos} valores únicos)")
    
    
```
